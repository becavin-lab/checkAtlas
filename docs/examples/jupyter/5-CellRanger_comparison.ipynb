# CheckAtlas examples : cellranger version

In this example, we download different cellranger files and try to load them in checkatlas

WARNING: Install and run only MultiQC from https://github.com/becavin-lab/MultiQC/tree/checkatlas. Otherwise checkatlas files will not be taken into account.

## Download cellranger dataset : PBMC 3k and PBMC 68k

The Cellranger file is directly downloaded from 10xGenomics database.

PBMC 6k chemistry v1 cellranger v1.1
https://www.10xgenomics.com/resources/datasets/pbmc-from-a-healthy-donor-no-cell-sorting-3-k-1-standard-2-0-0
Chemistry v1

PBMC 5k chemistry v3,1  cellranger v3.0.1
https://www.10xgenomics.com/resources/datasets/5-k-peripheral-blood-mononuclear-cells-pbm-cs-from-a-healthy-donor-v-3-chemistry-3-1-standard-3-0-2

PBMC 5k chemistry v3,1  cellranger v7.0.1
https://www.10xgenomics.com/resources/datasets/5-k-peripheral-blood-mononuclear-cells-pbm-cs-from-a-healthy-donor-v-3-chemistry-3-1-standard-3-0-2

PBMC 3k Multiome v1 ARC 2.0.0
https://www.10xgenomics.com/resources/datasets/pbmc-from-a-healthy-donor-no-cell-sorting-3-k-1-standard-2-0-0
Chemistry v1


```python
%%bash
mkdir -p data5
DATA_NAME=pbmc_6k_v1_v1
mkdir data5/${DATA_NAME}
mkdir data5/${DATA_NAME}/outs
cd data5/${DATA_NAME}/outs

curl -O https://cf.10xgenomics.com/samples/cell-exp/1.1.0/pbmc6k/pbmc6k_filtered_gene_bc_matrices.tar.gz
curl -O https://cf.10xgenomics.com/samples/cell-exp/1.1.0/pbmc6k/pbmc6k_analysis.tar.gz
curl -O https://cf.10xgenomics.com/samples/cell-exp/1.1.0/pbmc6k/pbmc6k_web_summary.html

tar -xvf pbmc6k_filtered_gene_bc_matrices.tar.gz
tar -xvf pbmc6k_analysis.tar.gz
rm pbmc6k_filtered_gene_bc_matrices.tar.gz
rm pbmc6k_analysis.tar.gz
```

```python
%%bash
mkdir -p data5
DATA_NAME=pbmc_5k_v3_v3
mkdir data5/${DATA_NAME}
mkdir data5/${DATA_NAME}/outs
cd data5/${DATA_NAME}/outs

curl -O https://cf.10xgenomics.com/samples/cell-exp/3.0.2/5k_pbmc_v3/5k_pbmc_v3_filtered_feature_bc_matrix.h5
curl -O https://cf.10xgenomics.com/samples/cell-exp/3.0.2/5k_pbmc_v3/5k_pbmc_v3_analysis.tar.gz
curl -O https://cf.10xgenomics.com/samples/cell-exp/3.0.2/5k_pbmc_v3/5k_pbmc_v3_web_summary.html

tar -xvf 5k_pbmc_v3_analysis.tar.gz
rm 5k_pbmc_v3_analysis.tar.gz
```

```python
%%bash
mkdir -p data5
DATA_NAME=pbmc_5k_v3_v7
mkdir data5/${DATA_NAME}
mkdir data5/${DATA_NAME}/outs
cd data5/${DATA_NAME}/outs

curl -O https://cf.10xgenomics.com/samples/cell-exp/7.0.1/SC3pv3_GEX_Human_PBMC/SC3pv3_GEX_Human_PBMC_filtered_feature_bc_matrix.h5
curl -O https://cf.10xgenomics.com/samples/cell-exp/7.0.1/SC3pv3_GEX_Human_PBMC/SC3pv3_GEX_Human_PBMC_analysis.tar.gz
curl -O https://cf.10xgenomics.com/samples/cell-exp/7.0.1/SC3pv3_GEX_Human_PBMC/SC3pv3_GEX_Human_PBMC_web_summary.html
tar -xvf SC3pv3_GEX_Human_PBMC_analysis.tar.gz
rm SC3pv3_GEX_Human_PBMC_analysis.tar.gz
```

```python
%%bash
mkdir -p data5
DATA_NAME=pbmc_3k_multiome
mkdir data5/${DATA_NAME}
mkdir data5/${DATA_NAME}/outs
cd data5/${DATA_NAME}/outs

curl -O https://cf.10xgenomics.com/samples/cell-arc/2.0.0/pbmc_unsorted_3k/pbmc_unsorted_3k_analysis.tar.gz
curl -O https://cf.10xgenomics.com/samples/cell-arc/2.0.0/pbmc_unsorted_3k/pbmc_unsorted_3k_web_summary.html
curl -O https://cf.10xgenomics.com/samples/cell-arc/2.0.0/pbmc_unsorted_3k/pbmc_unsorted_3k_filtered_feature_bc_matrix.h5
tar -xvf pbmc_unsorted_3k_analysis.tar.gz
rm pbmc_unsorted_3k_analysis.tar.gz
```

## Run checkatlas

If checkatlas is installed in your environment, you just need to run this cell. This will produce all metric tables and figures needed.

```python
%%bash
nextflow run nf-core-checkatlas -r dev --path /data/analysis/data_becavin/checkatlas_test/tuto/data5/
```
